```python
class LogisticRegression(nn.Module):
  def __init__(self, insize):
      super().__init__()
      self.log_reg = nn.Sequential(
      nn.Linear(insize,1),
      nn.Sigmoid()
    )
  def forward(self,x):
    return self.log_reg(x)
```
